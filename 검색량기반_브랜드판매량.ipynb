## Import

```python
import pandas as pd
import numpy as np
from tqdm.auto import tqdm
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

### 데이터 불러오기

```python
train_data = pd.read_csv('./train.csv').drop(['ID','제품','대분류','중분류','소분류'], axis=1)
```

```python
brand_data=pd.read_csv('./brand_keyword_cnt.csv')
```

```python
train_data.head()
```

```python
train_data = train_data.groupby('브랜드').sum()
train_data.head()
```

```python
brand_data=brand_data.set_index('브랜드').head()
```

```python
brand_data.head()
```

```python
data1_X=brand_data.iloc[0,:].transpose()   
```

```python
data1_y=train_data.iloc[0,:].transpose()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data1_X, data1_y, test_size=0.2, random_state=0)
```

```python
model=XGBRegressor()
model.fit(X_train, y_train)
pred=model.predict(X_test)
```

```python
print(mean_squared_error(y_test, pred))
```
